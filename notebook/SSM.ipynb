```python
import torch
import torch.nn as nn
import math
```

```python
def make_Diag_Block(Ab, l, L):
###対角ブロック行列
  def block(a, l, L):
    theta = 2*math.pi*a*l/(L+1)
    return torch.tensor([[math.cos(theta), -math.sin(theta)],
                         [math.sin(theta), math.cos(theta)]])
  Ab = torch.stack((Ab.real, Ab.imag), dim=1).flatten()
  return torch.block_diag(*[block(a, l, L) for a in Ab])

def Kernel_DiagBlock(A, P, Q, B, C, step, L):
  return torch.stack([C @ make_Diag_Block(A, l, L) @ B for l in range(L)])

make_Diag_Block(torch.randn(1, dtype=torch.cfloat), 1, 1)
```
